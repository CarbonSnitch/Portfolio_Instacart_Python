# Check whether there’s a difference in expenditure (the “prices” column) depending on the hour of the day. (Hint: To check this, you need to use an accurate sample for your line chart!)

```python
#Make smaller data so less RAM
df_3 = df_instacart [['prices', 'order_hour_of_day']]
df_3
```

```python
# Create a list for True/False for 70% 30% sample, use small sample
dev = np.random.rand(len(df_3)) <=0.7
big = df_3[dev]
small = df_3[~dev]
```

```python
len(df_3)
```

```python
len(big) + len(small)
```

```python
df_3_group = small
```

```python
#Make the group to make the graph from
df_3_group = df_3_group.groupby('order_hour_of_day').sum()
df_3_group
```

```python
#Make LIne Graph
line = sns.lineplot (data = df_3_group, x = 'order_hour_of_day', y = 'prices')
line.set_xlabel('Hour of Day')
line.set_ylabel('Sum Spent')
line.set_title('Instacart sum of sales per hour-of-day')
```

```python
#Export line graph
line.figure.savefig(os.path.join(path, '04 Analysis', 'Visualizations', 'Sales_per_hod.png'))
```

# You’ll also need to explore whether there’s a connection between age and spending power (income). To visualize this relationship, create a scatterplot using the sns.scatterplot() function.
# In a markdown cell beneath your scatterplot, describe what the scatterplot shows.

```python
df_5 = df_instacart [['Age', 'income']]
df_5
```

```python
df_5_group = df_5.groupby('Age').mean('income')
df_5_group
```

```python
scatter = sns.scatterplot (x = 'Age', y = 'income', data = df_5_group)
scatter.set_ylabel('Mean Income')
```

This scatterplot shows an unbelievable increase in income after the age of 40 where under the age of 40 there is an mean annual income of around 70,000, then at 41 the mean income jumps to about 116,000 where it remains stable until the age of 80.
